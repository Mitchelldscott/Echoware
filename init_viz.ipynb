```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_starfish(ax, l, alpha):
    offsets = np.linspace(0,2*np.pi,6)
    for a in offsets[:-1]:
        ax.plot([0, l*np.cos(a)*np.cos(alpha)], [0, l*np.sin(a)*np.cos(alpha)], [0, l*np.sin(alpha)])
    
```

```python
ax = plt.figure().add_subplot(projection='3d')
draw_starfish(ax, 1, 0)
plt.show()
```
